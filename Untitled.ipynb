```python
def evaluate_deployments(deployment_strings):
    error = 0
    success =0
    fail = 0
    for string in deployment_strings:
        item = eval(string)
        try:
            id = item["deployment_id"]
            if id[0:2] == 'd-'and id[2:].islower() and id[2:].isalnum() and len(id[2:]) == 10:
                if item["status"] == "Success":
                    success += 1
                elif item["status"] == "Fail":
                    fail += 1
                else:
                    error += 1
            else:
                error += 1
        except:
         error += 1
    return(success, fail, error)
```

```python
evaluate_deployments(['{"deployment_id":  "d-123456abcd", "status": "Fail"}'])
```
